# 함수의 리뷰

```python
import random

# 함수정의
def getNum():
    r=random.randint(1,45)
    return r

# 메인코드
lotto=[]

print('**로또 프로그램을 시작합니다.**')

while True:
    n=getNum()
    if lotto.count(n)==0:
        lotto.append(n)
    if len(lotto)==6:break

print('추첨된 로또 번호 : ', end= '')
lotto.sort()
#print(lotto)
for i in range(6):
    print(f'{lotto[i]} ', end='')
```

```python
import random

# 함수정의
def getNum():
    r=random.randint(1,45)
    return r

# 메인코드
lotto=list()
print('**로또 프로그램을 시작합니다.**')
while True:
    n=getNum()
    if lotto.count(n)==0:
        lotto.append(n)
    if len(lotto)==6:break

print('추첨된 로또 번호 : ', end= '')
lotto.sort()
print(lotto)
while True:
  n=getNum()
  if lotto.count(n)==0:break

for i in range(6):
    print(f'{lotto[i]} ', end='')
print(f'bonus {n}')
```

```python
from collections import Counter
filename='/content/drive/MyDrive/data/data2.txt'
f=open(filename)
#print(f.read())

wlist=f.read().split()
#print(wlist)

word=Counter(wlist) # **

print(word)

for w in word:
    print(f'{w}:{word[w]}')
print()

find_max=word.most_common(1) # **
print(find_max)

print('빈도수가 가장 높은 단어')
print(f'{find_max[0][0]} : {find_max[0][1]}')
```

```python
# 모듈 불러오기
from wordcloud import WordCloud
import matplotlib.pyplot as plt


filename='/content/drive/MyDrive/data/data2.txt'
f=open(filename)
data=f.read()
# 워드클라우드 객체 생성시 여러가지 속성들
wordcloud = WordCloud(
        background_color='pink',
        max_words = 200
        )

wordcloud = wordcloud.generate(data)
plt.figure(figsize=(15,15))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
